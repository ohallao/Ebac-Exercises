````python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Base_dados.csv')

df = df.drop(columns='Unnamed: 0', axis=1)
df
````

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
sns.pointplot(df, x= 'posse_de_imovel', y='renda')
````

````python
sns.pointplot(df, x= 'posse_de_veiculo', y='renda')
````

````python
sns.pointplot(df, x= 'posse_de_veiculo', y='renda', hue='tipo_renda', dodge= True)
````

````python
sns.pointplot(df, x= 'posse_de_imovel', y='renda', hue='tipo_renda', dodge= True)
````

Nos dois primeiros gráficos, as variáveis qualitativas discriminam a renda de forma estatisticamente significante. No entanto, percebe-se que a variável 'posse_de_veiculo' mostra-se como um fator mais determinante para predizer a renda, devido ao fato de que seus intervalos estão mais distantes dos que não possuem veículo e dos que possuem veículo.


Agora, se discriminarmos os dois gráficos pela variável "tipo_renda", observa-se uma significância da renda no perfil pensionista para a variável "posse_de_veiculo". Além disso, na variável "posse_de_imovel", verificou-se uma relevância da renda no perfil servidor público.

E não podemos deixar de mencionar que, de acordo com o gráfico, todos os bolsistas possuem imóvel, porém não possuem veículo, o que explica um intervalo tão alto em seu perfil.
